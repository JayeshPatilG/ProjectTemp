```python
import random
import pandas as pd

# Define the Individual class to represent each entry
class Individual:
    def __init__(self, row):
        self.id = row['ID']
        self.customer_id = row['Customer_ID']
        self.name = row['Name']
        self.age = row['Age']
        self.annual_income = row['Annual_Income']
        self.credit_score = row['Credit_Score']

    def calculate_credit_score(self):
        # Simplistic credit scoring based on annual income
        if self.annual_income > 50000:
            return 800
        else:
            return 500

# Generate a dataset with 100,000 entries (for demonstration, we'll generate random data)
data = []
for i in range(100000):
    entry = {
        'ID': i + 1,
        'Customer_ID': f'Cust{i + 1}',
        'Name': f'Individual{i + 1}',
        'Age': random.randint(20, 65),
        'Annual_Income': random.randint(20000, 120000),
        'Credit_Score': 0  # Placeholder for credit score
    }
    data.append(entry)

# Create a DataFrame from the generated data
df = pd.DataFrame(data)

# Calculate credit scores for each individual based on annual income
df['Credit_Score'] = df.apply(lambda row: row['Credit_Score'] if row['Annual_Income'] <= 50000 else 800, axis=1)

# Calculate average credit score
average_credit_score = df['Credit_Score'].mean()

# Display summary
print("Summary of Financial Status Analysis:")
print("------------------------------")
print(f"Total individuals: {len(df)}")
print(f"Average Credit Score: {average_credit_score:.2f}")
print(f"Number of individuals with good credit (score 800): {len(df[df['Credit_Score'] == 800])}")
```
